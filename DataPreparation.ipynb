So, In this project we will use some public books from https://gutenberg.org/ as our custom data set. 

##### Data preparation #####

```python
import torch;
```

```python
with open('data/pg36.txt' ,'r' , encoding = 'UTF-8') as f:
    text = f.read();
```

```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding(encoding_name="gpt2")
```

```python
text_encoding = tokenizer.encode(text)
```

```python
import numpy as np
```

```python
from torch.utils.data import Dataset, DataLoader

class DataSet(Dataset):
    def __init__(self , features :np.array , target:np.array) -> None:
        super().__init__();
        self.x_data = features;
        self.y_data = target;
    
    
    def __getitem__(self, index):
        # print(self.x_data);
        x_data_ = torch.tensor(self.x_data[index]);
        y_data_ = torch.tensor(self.y_data[index]);
        return x_data_ , y_data_;

    def __len__(self):
        return len(self.x_data);


def create_data_set(text_encoding, stride = 4 , batch_size = 8) -> DataLoader:
    features = np.array([text_encoding[i:i+stride]for i in range(0,len(text_encoding),stride)][:-2]);
    target =  np.array([text_encoding[i:i+stride]for i in range(1,len(text_encoding),stride)][:-2]);
    dataSet =  DataSet(features,target);
    dataLoader = DataLoader(dataSet , batch_size=batch_size);
    return dataLoader;
    

    
```

```python
dataLoader = create_data_set(text_encoding);
for item in dataLoader:
    print(item)
```
